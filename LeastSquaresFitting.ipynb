### **This Section imports all needed packages and plots the given measurements.**
### Run this section first.

```
# import needed packages
import numpy as np
import matplotlib.pyplot as plt

# create x & y arrays
x = np.array([0.6, 0.7, 0.8, 0.9, 1.0])
y = np.array([1.01, 0.81, 0.70, 0.65, 0.62])

# plot measurements
plt.scatter(x, y)
plt.xlabel('Illumination [10³ lux]')
plt.ylabel('Resistance R [kOhm]')
plt.title('Measurements')
plt.show()
```

In this section we create the Matrix X for polyn. m=2 and calculate polynomial coefficients.
Then we plot the fitted graph.

```
# matrix X for polyn. degree m=2
X = [np.power(x,2), np.power(x, 1), np.power(x, 0)]
X = np.array(X)
X = np.transpose(X)

# calculate the prseudoinverse
Xpinv = np.linalg.pinv(X)

# calculate the polynomial coefficients
c = Xpinv.dot(y)

# create dense x array for plotting purposes
xdense = np.arange(0.4, 1.201, 0.001)

# evalualte polynomial at dense x values
ypoly = np.polyval(c, xdense)

# plot fitted graph
plt.scatter(x,y)
plt.plot(xdense, ypoly)
plt.xlabel('Illumination [10³ lux]')
plt.ylabel('Resistance R [kOhm]')
plt.title('Fitted graph')
plt.show()
```
